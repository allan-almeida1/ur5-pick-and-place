```python
from ur5 import get_dataset
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
folder = 'machine_learning/img/'
data = get_dataset(folder)
r = np.random.randint(0,4999)
x_img = data[r]['x_img']
y_img = data[r]['y_img']
img = plt.imread(folder + data[r]['raw_file'])
img1 = img.copy()
cv2.circle(img1, (int(x_img), int(y_img)), 3, (0, 0, 1), -1)
plt.imshow(img1)
plt.show()
```
